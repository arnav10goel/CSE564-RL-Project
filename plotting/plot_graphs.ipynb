```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("00_bigfish_episodic_return.csv")
df = df.dropna()

steps = df["Step"]
episodic_return = df["0_sazut-tonat-dokabtofok - episodic_return"]
episodic_return_min = df["0_sazut-tonat-dokabtofok - episodic_return__MIN"]
episodic_return_max = df["0_sazut-tonat-dokabtofok - episodic_return__MAX"]

# Apply EMA-based smoothing
ema_episodic_return = episodic_return.ewm(span=250, adjust=False).mean()

plt.plot(steps, ema_episodic_return, label="PPO from LAPO latent policy")
plt.fill_between(steps, episodic_return_min, episodic_return_max, alpha=0.2)

plt.axhline(y=urp_ep_return["bigfish"], color="r", linestyle="--", label="Uniform Random Policy")
plt.axhline(y=expert_ep_return["bigfish"], color="g", linestyle="--", label="Expert PPO Policy")

plt.xlabel("Steps")
plt.ylabel("Episodic Return")
plt.yticks(np.arange(0, 50, 5))
plt.title("Episodic Return for Bigfish")
plt.grid()
plt.legend()
plt.show()
```
